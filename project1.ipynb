### now we use the mean manupulation method so that we can replace null values with mean value
#### Mean manupulation method refers to the method in which we replace the null value with mean value 

```python
#inspecting the null values in the dataset
#print(df.isnull().values.sum())
print(df.isnull().values.sum())

#replace the ' 's with NaN
df = df.replace(" ",np.NaN)
#df=df.replace(" ",np.NaN)
#impute the missing values with mean imputation
df=df.fillna(df.mean())
#count the numbers again for verification
print(df.isnull().values.sum())
```

```python
#seaborn histogram
sns.displot(data=df, x="price", kind="hist")
#add labels
plt.title('Price View according to the count of the car')
plt.xlabel('Price')
plt.ylabel('Count')
```

# Interpretation

>* Here price of the car data is shown in x axis whereas the count is shown in y axis. 
>* Cars of around 10k to 20k is found be sold maximum whereas cars around 40k to 50k are found to be sold minumum according to the graph.  

```python
sns.displot(data=df, x="mileage",kind="kde")
#sns.displot(data=df, x="absences", kind="hist")

plt.title('Mileage given by car to the count')
plt.xlabel("Mileage")
plt.ylabel("count")

#show the graph
plt.show()
```

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(16, 12))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
plt.figure(figsize=(20,20))
plt.title('Pearson Correlation of Features', size = 15)
colormap = sns.diverging_palette(10, 220, as_cmap = True)
sns.heatmap(df.corr(),
            cmap = colormap,
            square = True,
            annot = True,
            linewidths=0.1,vmax=1.0, linecolor='white',
            annot_kws={'fontsize':20 })
plt.show()
```

# Intrepretation of the graph
>* The relationship between the components are shown according to the color.
>* -0.2 means the given features are not so related whereas 1 shows that they are highly related to each other and 0.2 and above is also considered as highly related.
>* The graph shows that price and mileage are not so related.
